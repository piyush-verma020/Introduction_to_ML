## Train Test Split

**Train-test split** is a fundamental technique in **supervised machine learning** used to evaluate how well a model can generalise to new, unseen data. Before training, a model is like a child with no knowledge; the split provides a structured way to teach the model and then test its understanding.

### Why the need of Splitting Data
The primary reason for splitting data is to prevent **overfitting and memorisation**. 
*   **Avoiding Memorisation:** If a model is trained and evaluated on 100% of the same dataset, it will simply **memorise the relationships** between features and labels. This often leads to a deceptive **perfect prediction score of 1.0 (100% accuracy)**, which fails in real-world scenarios when the model encounters new data [source paragraph from history, 289].
*   **Generalisation:** By setting aside a portion of data that the model never sees during training, you can judge its **real-world performance**. If the model performs well on the training set but poorly on the test set, it indicates the model has not learned the underlying concepts correctly.

### How the Split is Implemented
In Python, this is typically handled using the **`train_test_split`** function from the `sklearn.model_selection` library. The process involves several key steps and parameters:

*   **Input and Output Separation:** First, the dataset is divided into **independent variables (X)** and the **dependent/target variable (y)**.
*   **Generating Four Sets:** The function returns four distinct datasets: **X_train, X_test, y_train, and y_test**.
*   **Test Size:** You must decide what percentage of data to reserve for testing. Common ratios are **20% or 25%** (`test_size=0.2` or `0.25`), though this can depend on the total size of the dataset.
*   **Random State:** Using a parameter like **`random_state=42`** ensures that the data is split in the same way every time the code is run. This consistency is crucial for comparing results when trying to improve model accuracy.

### Workflow in Model Building
1.  **Training Phase:** The **X_train and y_train** sets are passed into the model’s **`.fit()`** method to "teach" the algorithm the patterns in the data.
2.  **Testing Phase:** The **X_test** set is then used with the **`.predict()`** or **`.score()`** methods to see how the model behaves with data it has never encountered before. 
3.  **Refinement:** If the resulting accuracy (score) is poor, the model can be retrained by adjusting hyperparameters or changing the `random_state` to find a more effective training split.

### NOTE:

**Why should scaling be performed after train-test splitting?**

Train-test splitting and feature scaling are two different steps in machine learning preprocessing.

* **Train-test split** divides the dataset into training data and testing data.
    * The training data is used to train the model.
    * The testing data is used to evaluate model performance on unseen data.
* **Feature scaling** changes the magnitude of feature values so that machine learning algorithms can learn more efficiently and converge faster.

Scaling should not be applied before splitting the dataset because it can lead to **data leakage**. If scaling is performed on the entire dataset first, the scaler learns information from both the training and testing data. As a result, the testing data indirectly influences the training process, making the model evaluation unrealistic.

Therefore, the correct workflow is:

1. Split the dataset into training and testing sets.
2. Fit the scaler only on the training data.
3. Transform the training data using `fit_transform()`.
4. Transform the testing data using `transform()` using the same scaler.

This ensures that the model is evaluated fairly on completely unseen data.

# Regression Analysis

**Regression analysis** is a type of supervised machine learning used when the target output is **continuous in nature**, meaning it follows a sequence of real numbers rather than discrete categories. It is primarily used to predict future values based on historical data patterns.

### Types of Regression Algorithms
Regression techniques are generally categorized into two groups based on the relationship between input and output features:

*   **Linear Algorithms:** These are used when there is a linear relationship between inputs and outputs. 
    *   **Simple Linear Regression:** Used when there is only **one independent input feature** to predict an output. It follows the line equation **$y = mx + c$**, where $m$ is the slope and $c$ is the intercept.
    *   **Multiple Linear Regression:** Used when there are **multiple input features** predicting a single output. Instead of a line, this model creates a **plane or hyper-plane** using the equation $y = m_1x_1 + m_2x_2 + \dots + c$.
    *   **Regularized Regression:** Includes **Lasso (L1)** and **Ridge (L2)**. Lasso can shrink unimportant coefficients to zero for **feature selection**, while Ridge reduces coefficient values to prevent **overfitting**.
*   **Non-Linear Algorithms:** Used when the data follows a curved or random pattern rather than a straight line.
    *   **Polynomial Regression:** Adds power terms (like $x^2$ or $x^3$) to the features to fit a curved "best fit" line.
    *   **Others:** Decision Trees, Random Forest, Support Vector Machines (SVM), and K-Nearest Neighbors (KNN) can also be adapted for regression tasks.

### Key Concepts in Model Building
*   **The Best Fit Line:** The goal of regression is to find a line or plane that minimizes the **error (loss)** between the actual data points and the predicted values.
*   **Cost Functions:** These mathematical formulas measure how "wrong" a model is. Common regression cost functions include:
    *   **Mean Squared Error (MSE):** Squares the errors; it is useful for optimization but sensitive to outliers.
    *   **Mean Absolute Error (MAE):** Takes the absolute difference; it is more robust when outliers are present in the data.
    *   **Root Mean Squared Error (RMSE):** The square root of MSE, which returns the error to the original unit of measurement.
*   **Gradient Descent:** This is an optimization technique used to iteratively update the model's parameters ($m$ and $c$) to reach the point where the **cost function is at its minimum**.
*   **Data Preparation:** Effective regression often requires **Feature Scaling** (Standardization or Normalization) to ensure large numerical values don't dominate the model, and **Outlier Detection** to prevent distorted predictions.

### Evaluation and Accuracy
To determine how well a regression model performs on unseen data, the dataset is typically split into **training and testing sets**. The model's performance is often measured using a **score** (like the R2 score), where a value closer to 1.0 indicates higher accuracy. However, a perfect score of 1.0 often warns of **overfitting**, where the model has simply memorised the training data instead of learning general patterns.
