### Key difference between Z-score method and Windorization

Both Winsorization and the Z-score method deal with outliers and are often used during data preprocessing to reduce their impact.

**Z-score method**

* Detects outliers

* Based on how many standard deviations a value is from the mean

* Formula:

            $$z = \frac{x - \mu}{\sigma}$$

* Typical rule: values with ∣z∣>3 are considered outliers

* What happens to outliers?

    * Usually removed or flagged

    * Sometimes capped, but that’s not the core idea


**Winsorization**

* Modifies outliers instead of removing them

* Extreme values are replaced with boundary values (percentiles)

* Example:

    * Bottom 1% → replaced with 1st percentile value

    * Top 1% → replaced with 99th percentile value

* Keeps dataset size the same

* Reduces the influence of extreme values on mean & variance

### Handling Duplicate Data

Removing duplicate data from a dataset is essential because duplicate entries can negatively impact the performance of machine learning algorithms. If not addressed, duplicates may introduce bias, distort model training, and lead to inaccurate predictions. Additionally, redundant data increases the overall dataset size unnecessarily, which can slow down processing time and reduce computational efficiency.



```python
import pandas as pd
```

```python
data = {"name": ["a","b","c","d","a","c"], "eng":[8, 7, 5, 8, 8, 4], "maths": [2, 3, 4, 5, 2, 4]}    # Here, a contains duplicate data but c doesn't
```

```python
# Let's create a dataset
df = pd.DataFrame(data)
```

```python
df
```

```python
df.duplicated()
```

Here, we can clearly observe that the entries at index 0 and 4 contain identical values, whereas the entries at index 2 and 5 differ from each other. Therefore, duplicate data is present only at indices 0 and 4.

```python
# Let's drop the duplicate columns
df.drop_duplicates(keep='first', inplace=True)
```

The argument `keep='first'` lets you keep the first duplicate value and removes the rest

### Checking vs Removing Duplicates

It is better to **check for duplicate values** first instead of removing them immediately. This helps us understand whether duplicates actually exist before modifying the dataset.

Using:

***loan_df.duplicated().unique()***


is a safe way to check if duplicates are present. Since `.duplicated()` returns boolean values (`True` for duplicates and `False` for unique rows), if `True` does not appear, it means there are no duplicate rows.

However, a more practical way to check duplicates is:

***loan_df.duplicated().sum()***


This directly tells us the number of duplicate rows:

* If the result is `0`, there are no duplicates.

* If the result is greater than `0`, duplicates are present.

After confirming duplicates exist, we can remove them using:

***loan_df.drop_duplicates(keep='first', inplace=True)***


Checking first is considered a good data practice because it avoids modifying the dataset blindly and helps maintain transparency in the data cleaning process.

## CoapplicantIncome – Outliers, Z-score & Winsorization Summary

#### Distribution Diagnosis

From the statistics:

* Mean ≈ 1621

* Std ≈ 2926 (very high)

* 25% = 0

* Median = 1188

* 75% = 2297

* Max = 41667

Interpretation:

* Many values are 0 (no co-applicant)

* A few very large incomes stretch the mean

* Distribution is highly right-skewed

* This is not normally distributed

* This is structural skew, not noise.

#### Why Z-Score Removed 67%

Z-score assumes:

##### *Data ∼ Normal Distribution*

But your data is:

* Zero-inflated

* Right-skewed

* Heavy-tailed

* So when applying:

  **∣𝑧∣>3**

It incorrectly flags many normal values as outliers.

Problem is **misuse of Z-score**, not the data itself.

#### Winsorization Concern Clarified

Winsorization:

* Caps only extreme percentiles

* Does NOT replace all large values with one single number

* Only affects the chosen tail %

Example:

* Winsorize at 95%

* Only top 5% capped

* 95% unchanged

So 67% will NOT collapse into the same value unless you cap 67%.

#### When Winsorization Works Well

* Rare but extreme anomalies
* Symmetric distribution with few tail errors
* Linear models sensitive to extreme values

***Winsorization = robustness tool***

#### When Winsorization Is NOT Ideal

* Highly skewed data
* Zero-inflated distributions
* Financial features with natural long tails

It compresses real signal instead of fixing noise.

### Better Approach for CoapplicantIncome

* Log Transform (Best choice)

        np.log1p(CoapplicantIncome)

Why?

* Compresses large values

* Preserves ranking

* Keeps variance structure

* No artificial duplicates

#### Or leave as-it-is for tree models

Models like:

* Random Forest

* XGBoost

* LightGBM

are naturally robust to skewness.

#### Key Takeaways

* High variance ≠ bad feature

* Z-score fails on skewed data

* Winsorization is for extreme anomalies, not skew correction

* Log transform > Winsorization for skewed financial features

## Final Understanding



Winsorization is useful for extreme anomalies, but not appropriate for naturally skewed financial features like CoapplicantIncome.
